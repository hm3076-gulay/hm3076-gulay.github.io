## **Step 2: Mean / Median / Mode (Pandas)**

```python
# Using the numeric column prices.amountMin
mean_price = df["prices.amountMin"].mean()
median_price = df["prices.amountMin"].median()
mode_price = df["prices.amountMin"].mode()[0]   # mode() returns a Series

# Put results into a small table
stats_table = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value ($)": [round(mean_price, 2), round(median_price, 2), round(mode_price, 2)]
})

stats_table
```

**Interpretation:** Mean reflects the average price (sensitive to outliers), the median is the “middle”
typical price, and the mode is the most common price point in the dataset.

## **Step 3: Pure Python (The Hard Way)**

```python
import csv

values = []

with open("womenshoeprices.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        val_str = row["prices.amountMin"]
        if val_str != "" and val_str is not None:
            try:
                val = float(val_str)
                values.append(val)
            except:
                pass

# Mean (hard way)
mean_hard = sum(values) / len(values)

# Median (hard way)
sorted_vals = sorted(values)
mid = len(sorted_vals) // 2

if len(sorted_vals) % 2 == 1:
    median_hard = sorted_vals[mid]
else:
    median_hard = (sorted_vals[mid - 1] + sorted_vals[mid]) / 2

# Mode (hard way)
counts = {}
for v in values:
    counts[v] = counts.get(v, 0) + 1

max_count = max(counts.values())
mode_hard = [v for v, c in counts.items() if c == max_count]

print(mean_hard)
print(median_hard)
print(mode_hard)
```

_These values match the pandas results (within rounding), confirming the manual implementation._

## **Step 4: Visualization (standard library only)**

```python
# Showing 15 shoes with text bars for price
sample = df.sample(15)

for i, row in sample.iterrows():
    price = row["prices.amountMin"]
    if pd.notna(price):
        bar = "$" * int(price // 10)   # one $ per $10
        print(f"{row['brand']}: {bar}")
```

**How to read this.** Each `$` symbol represents approximately **$${DOLLARS_PER_SYMBOL}**.
More `$` = higher price. 

This text chart conveys relative prices without any plotting libraries.
